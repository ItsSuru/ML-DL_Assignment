#### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
%matplotlib inline
```

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris() 
reduced = PCA(n_components=3).fit_transform(iris.data) 
Y = iris.target
```

```python
fig = plt.figure(1, figsize=(15, 8)) 
ax = Axes3D(fig, elev=-150, azim=110)

ax.scatter(reduced[:, 0], reduced[:, 1], reduced[:, 2], c=Y, cmap=plt.cm.rainbow_r)

ax.set_title("First Three PCA Directions") 
ax.set_xlabel("1st eigen_vector") 
ax.set_ylabel("2nd eigen_vector") 
ax.set_zlabel("3rd eigen_vector")

plt.show()
```
